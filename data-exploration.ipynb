```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("dylanjcastillo/7k-books-with-metadata")

print("Path to dataset files:", path)
```

```python
import pandas as pd
```

```python
books = pd.read_csv(f"{path}/books.csv")
```

```python
books.head(n=1)
```

```python
books.shape
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
ax = plt.axes()
sns.heatmap(books.isna().transpose(),cbar=False,ax=ax)
plt.xlabel("Column")
plt.ylabel("Missing Values")
plt.show()
```

```python
import numpy as np

books["missing_description"] = np.where(books["description"].isna(), 1, 0)
books["age_of_book"] = 2025 - books["published_year"]
```

```python
books.head(n=1)
```

```python
columns_of_interest = ["num_pages", "age_of_book", "missing_description", "average_rating"]

correlation_matrix = books[columns_of_interest].corr(method="spearman")
# method="pearson" - linear relationship
# method="spearman" - monotonic relationship
sns.set_theme(style="white")

plt.figure(figsize=(8,6))
heatmap = sns.heatmap(
    correlation_matrix,
    annot=True,        # show correlation numbers
    fmt=".2f",         # format numbers with 2 decimal places
    cmap="coolwarm",   # red-blue color scale
    cbar_kws={"label": "Spearman correlation"}  # label colorbar
)


heatmap.set_title("Correlation heatmap")
plt.show()
```

```python
result = ~books["num_pages"].isna()
print(type(result)) # 'pandas.Series'
print(result.dtype)
```

```python
book_missing = books[~(books["description"].isna()) |
      (~books["num_pages"].isna()) |
      (~books["average_rating"].isna()) |
      (~books["published_year"].isna())
]
```

```python
book_missing = books.dropna(
    subset=["description", "num_pages", "average_rating", "published_year"],
    how="all"
)
```

```python
book_missing
```

```python
book_missing["categories"].value_counts().reset_index().sort_values("count", ascending=False)

```

```python
book_missing["words_in_description"] = book_missing["description"].str.split().str.len()
```

```python
book_missing
```

```python
book_missing["words_in_description"].describe()
```

```python
# .loc[row_condition, column_name]
book_missing.loc[
    book_missing["words_in_description"].between(1, 4),
    "description"
]
```

```python
book_missing.loc[
    book_missing["words_in_description"].between(25, 34),
    "description"
]
```

```python
book_missing_25_words = book_missing[book_missing["words_in_description"] >= 25]
```

```python
book_missing_25_words
```

```python
book_missing_25_words.shape
```

```python
book_missing_25_words["title_and_subtitle"] = np.where(
    book_missing_25_words["subtitle"].notna(),
    book_missing_25_words["title"] + ": " + book_missing_25_words["subtitle"],
    book_missing_25_words["title"]
)
```

```python
book_missing_25_words
```

```python
book_missing_25_words["tagged_description"] = np.where(
    book_missing_25_words["description"].notna(),
    book_missing_25_words["isbn13"].astype(str) + ": " + book_missing_25_words["description"].astype(str),
    book_missing_25_words["isbn13"].astype(str)
)
```

```python
book_missing_25_words.head(n=1)
```

```python
book_missing_25_words.drop(
    ["subtitle", "missing_description", "age_of_book", "words_in_description"],
    axis=1
).to_csv("books_cleaned.csv", index=False)
```
